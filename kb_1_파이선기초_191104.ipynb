## 쥬피터 팁

- 각 셀은 프로그램 코드 또는 문서(마크업 문서 지원) 둘 중 하나로 사용된다.
- 셀을 실행하려면 Shift + Enter 를 입력한다 (셀 실행후 커서가 다음 셀로 이동)
- 셀을 Ctrl + Enter로 실행하면 셀 실행 후에 현재의 셀에 남아 있는다. 
### 코랩에서 셀에 대한 명령어 (ctrl + m + ...)
- h  help
- a  위에 셀 추가 (above)
- b  아래에 셀 추가 (below)
- d  셀 삭제 (delete)
- m  문서 모드로 전환 (mark up)
- y  code 모드로 전환
- o  출력 보이기/안보이기 (output)
- '-' 셀 둘로 나누기
- 'shift'+클릭으로 두 개의 셀 선택 후에 (ctrl + shift + M) 셀 합치기

### PC에서 사용할 때
-  편집창 아닌 명령 모드(셀 외부를 클릭) 에서 ctrl + m 없이 단축 문자만 입력 (a, b, x, o 등)
-  셀을 나눌 때는 ctrl + shift + '-'
-  바로 아래의 셀과 합칠 때는 shift + m

## range, arange

```python
# 간단히 리스트를 만드는 방법
x = [i*10 for i in range(5)] 
x
```

# if 문
- if 문장의 끝에는 ':'가 있어야 하며 if 문이 적용되는 블록(block)은 반드시 들여쓰기 (indentation)을 해야 한다. 

```python
x=7
if x > 10:
    print("x > 10") 
    print('---------')
elif x > 3:
    print("10 > x > 3")
    print('---------') 
else:
    print(" x <= 3")
 
```

## for

## def (함수정의)

```python
def my_func(a, b, c): 
    return a*100 + b*10 + c
```

```python
print(my_func(1,2,3))
```

```python
def my_func(a=5, b=5, c=5): 
    return a*100 + b*10 + c
```

```python
print(my_func(1,2,3))
print(my_func(1,2))
print(my_func(1))
print(my_func())
```
